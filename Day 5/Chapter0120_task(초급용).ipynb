__[과제 1: 초기화 메서드 __init__]__

목표: __init__ 메서드를 구현하여 붕어빵 가게의 초기 재고, 가격, 판매량을 설정합니다.

설명: self.stock, self.prices, self.sales 딕셔너리를 초기화하세요.
- stock 
    - "팥붕어빵": 10
    - "슈크림붕어빵": 8
    - "초코붕어빵": 5

- prices
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- sales
    - "팥붕어빵": 0,
    - "슈크림붕어빵": 0,
    - "초코붕어빵": 0

```python
def __init__(self):
    self.stocks = {
        "팥붕어빵": 10,
        "슈크림붕어빵": 8, 
        "초코붕어빵": 5
    }

    self.prices = {
        "팥붕어빵": 1000,
        "슈크림붕어빵": 1200,
        "초코붕어빵": 1500
    }

    self.sales = {
        "팥붕어빵": 0,
        "슈크림붕어빵": 0,
        "초코붕어빵": 0
    }
    # 딕셔너리는 키를 통해 원하는 값을 빠르게 조회할 수 있는 자료구조임 
    # 처리 속도가 빠름 
    # 값을 키와 쌍으로 저장하는데 편리함 
    # 리스트나 튜플과는 달리 인덱스가 아닌 키로 데이터를 다루기 때문에 가독성이 좋아짐
    # init에 넣는 정보들: 
    # 1)이 객체의 상태를 나타내거나 (주소, 이름 등 바뀌지 않는 것들)
    # 2) 이 정보가 다른 메서드에서도 사용 되어야 하거나. 
    # 3) 이 정보가 객체가 생성될 때 한번만 설정 되거나. 
    # 매번 계산되는 값은 init에 넣지 않음 (임시변수 등)
```

__[과제 2: 재고 확인 메서드 check_stock]__

목표: check_stock 함수를 구현하여 현재 붕어빵 재고를 출력합니다.

설명: self.stock 딕셔너리를 순회하며 각 붕어빵의 종류와 재고를 출력하세요.


```python
def check_stock(self):
    print("현재 재고 현황")
    for cake_type, quantity in self.stocks.items():
        print(f"{cake_type}: {quantity}개")


# def check_Stock(self):
#     for cake_type in self.stock:
#         print(f"{cake_type}: {self.stock[cake_type]}개")
    
# self.stock.item()은 위의 self.stock의 모든 키-값 쌍을 하나씩 꺼내서 확인함
# .item()을 사용하면 키와 값이 한 번에 튜플로 묶여 반환됨  
# .item을 쓰는게 훨씬 더 좋음. 가독성이 좋아짐. 
```

__[과제 3: 주문 처리 메서드 process_order]__

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
def process_order(self, cake_type, cake_count):
    try:
        if self.stocks[cake_type] >= cake_count:
            self.stocks[cake_type] -= cake_count
            self.sales[cake_type] += cake_count 
            print(f"{cake_type}이 {cake_count}개 판매 되었습니다. 주문이 완료되었습니다.")
        else: 
            print(f"{cake_type} 재고가 충분하지 않습니다.")
    except KeyError:
        print("존재하지 않는 붕어빵입니다.") 

    # 1) bread_type이 self.stock에 존재하는지 확인하세요.
    #    - 없다면 "없는 메뉴입니다." 메시지를 출력하고 return 처리.
        
    # 2) 재고가 부족하면 "재고가 부족합니다." 메시지를 출력하고 return 처리.
        
    # 3) 재고가 충분하면:
    #    - 재고에서 주문량을 빼세요.
    #    - 판매량을 주문량만큼 더하세요.
    #    - 주문 완료 메시지를 출력하세요. 예: "주문이 완료되었습니다. {bread_type} {bread_count}개"
```

__[과제 4: 관리자 모드 메서드 admin_mode]__

목표: admin_mode 메서드를 구현하여 관리자 모드에서 재고를 추가할 수 있도록 합니다.

설명: bread_type과 additional_stock을 입력받아 재고를 추가합니다. 잘못된 입력 값이 들어온 경우 적절한 메시지를 출력하세요.

```python
# 매개변수는 함수가 외부에서 데이터를 받아올 때 사용하면 됨 
# admin_mode는 외부에서 데이터를 받아올 필요 없이 함수 내부에서 정보값을 입력 받음 
# 모든 데이터가 이미 객체 안에 있음. 

def admin_mode(self):
    print("관리자 모드입니다")
    while True:
        cake_type = input("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵) 종료를 원하시면 '종료'를 입력하세요: ")
        if cake_type == "종료":
            break
        try: 
            additional_stock = int(input(f"{cake_type}의 추가 재고량을 입력하세요:"))
            self.stocks[cake_type] += additional_stock
            print(f"{cake_type} 재고가 {additional_stock}개 추가되었습니다") 
            self.check_stock()
        except ValueError:
            print("숫자를 입력해주세요")
        except KeyError: 
            print("팥붕어빵, 슈크림붕어빵, 초코붕어빵을 입력하세요.")
```


__[과제 5: 매출 계산 메서드 calculate_total_sales]__

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
def calculate_total_sales(self):
    total_sales = 0
    total_sales += self.sales["팥붕어빵"] * self.prices["팥붕어빵"]
    total_sales += self.sales["슈크림붕어빵"] * self.prices["슈크림붕어빵"]
    total_sales += self.sales["초코붕어빵"] * self.prices["초코붕어빵"]
    return f"총 매출 {total_sales}원"
    # 1) total_sales 변수를 0으로 초기화하세요.
        
    # 2) self.sales 딕셔너리를 순회하면서:
    #    - 각 붕어빵의 판매량과 가격을 곱해 total_sales에 더하세요.
        
    # 3) 최종적으로 총 매출을 출력하세요. 예: "총 매출: {total_sales}원"
```

__[과제 7: 최종 BungeoppangShop 클래스]__

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다. 

```python
class BungeoppangShop:
    def __init__(self):
        self.stocks = {
        "팥붕어빵": 10,
        "슈크림붕어빵": 8, 
        "초코붕어빵": 5
    }

        self.prices = {
        "팥붕어빵": 1000,
        "슈크림붕어빵": 1200,
        "초코붕어빵": 1500
    }

        self.sales = {
        "팥붕어빵": 0,
        "슈크림붕어빵": 0,
        "초코붕어빵": 0
    }
    
    def check_stock(self):
        print("현재 재고 현황")
        for cake_type, quantity in self.stocks.items():
            print(f"{cake_type}: {quantity}개")

    def process_order(self, cake_type, cake_count):
        try:
            if self.stocks[cake_type] >= cake_count:
                self.stocks[cake_type] -= cake_count
                self.sales[cake_type] += cake_count 
                print(f"{cake_type}이 {cake_count}개 판매 되었습니다. 주문이 완료되었습니다.")
            else: 
                print(f"{cake_type} 재고가 충분하지 않습니다.")
        except KeyError:
            print("존재하지 않는 붕어빵입니다.") 
    def admin_mode(self):
        print("관리자 모드입니다")
        while True:
            cake_type = input("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵) 종료를 원하시면 '종료'를 입력하세요: ")
            if cake_type == "종료":
                break
            try: 
                additional_stock = int(input(f"{cake_type}의 추가 재고량을 입력하세요:"))
                self.stocks[cake_type] += additional_stock
                print(f"{cake_type} 재고가 {additional_stock}개 추가되었습니다") 
                self.check_stock()
            except ValueError:
                print("숫자를 입력해주세요")
            except KeyError: 
                print("팥붕어빵, 슈크림붕어빵, 초코붕어빵을 입력하세요.")
            
    def calculate_total_sales(self):
        total_sales = 0
        total_sales += self.sales["팥붕어빵"] * self.prices["팥붕어빵"]
        total_sales += self.sales["슈크림붕어빵"] * self.prices["슈크림붕어빵"]
        total_sales += self.sales["초코붕어빵"] * self.prices["초코붕어빵"]
        return f"총 매출 {total_sales}원"
    
def main():
    shop = BungeoppangShop()
    print("안녕하세요? 붕어빵 가게입니다. 붕어빵 그만드세요 힘들어요")

# 무한 루프 구문. 프로그램이 계속 실행 되도록 하는것. for, while 두 개가 잘 안써지고 어려움 .... ㅠ 
# 사용자가 종료를 선택해서 break를 누를 때까지 계속 반복 
    while True: 
        cake_order = input("주문, 관리자, 종료 중 원하는 메뉴를 입력하세요")
        if cake_order == "주문":
            cake_type = input("붕어빵 종류를 입력하세요: 팥붕어빵, 슈크림붕어빵, 초코붕어빵")
            cake_count = int(input("원하는 개수를 입력하세요:"))
            
            # 계속 shop.이 붙어야 하는 이유는 앞에서 만든 shop이라는 객체를 계속 불러오는 거임 
            # 객체 지향 프로그래밍의 핵심임 
            # shop은 가게 그 자체고, 가게에서 기능이 process_order라고 보면 되는거임. 
            shop.process_order(cake_type, cake_count)
        
            print("주문 후 재고 현황")
            shop.check_stock()
            
        elif cake_order == "관리자":
            shop.admin_mode()

        elif cake_order == "종료":
            total = shop.calculate_total_sales()
            print(total)
            print("주문이 종료 되었습니다. 감사합니다.")
            break
        else: 
            print("주문, 관리자, 종료 중 선택하세요.")

if __name__ == "__main__":
    main()

```
