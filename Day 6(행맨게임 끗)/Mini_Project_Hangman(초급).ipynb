ë¬¸ì œê°€ ì–´ë ¤ìš°ì‹  ê²½ìš° êµ¬ê¸€ë§ í†µí•´ ë¬¸ì œë¥¼ í•´ê²°í•´ë³´ì„¸ìš”.  

```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
import random

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# í–‰ë§¨ ê·¸ë¦¼
# """"""ëŠ” ì—¬ëŸ¬ ì¤„ ë¬¸ìì—´ì„ ë§Œë“œëŠ” ë°©ë²•ì„
# """"""ë¡œ ë¬¶ì¸ í•˜ë‚˜ì˜ ê·¸ë¦¼ì´ ì¸ë±ìŠ¤ 1ê°œë¡œ ì½í˜
hangman_pics = [
    # ì¸ë±ìŠ¤ 0ë²ˆ
    """------
     |    |
     |
     |
     |
     |
    ---""",
    # ì¸ë±ìŠ¤ 1ë²ˆ 
     """ ------
     |    |
     |    O
     |
     |
     |
    ---""",
    # ì¸ë±ìŠ¤ 2ë²ˆ
     """------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    # ì¸ë±ìŠ¤ 3ë²ˆ
     """------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    # ì¸ë±ìŠ¤ 4ë²ˆ 
     """------
     |    |
     |    O
     |   /|\
     |
     |
    --- """,
    # ì¸ë±ìŠ¤ 5ë²ˆ
     """------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    # ì¸ë±ìŠ¤ 6ë²ˆ
    """------
     |    |
     |    O
     |   /|\
     |   / \
     |
    --- """ 
    ]


class HangmanGame:
    def __init__(self, words):
        # ëœë¤í•˜ê²Œ ì„ íƒë  ë‹¨ì–´.
        self.word = random.choice(words).lower() 
        # ì‹œë„ ìˆ˜ 6ë²ˆ 
        # ì¶”ì¸¡í•œ ì•ŒíŒŒë²³ì„ ë‹´ì„ ë¹ˆ ì§‘í•©
        # ë¦¬ìŠ¤íŠ¸ë¡œë„ ê°€ëŠ¥í•˜ì§€ë§Œ ì¤‘ë³µ ì²´í¬ë¥¼ ìœ„í•´ í•¨ìˆ˜ ì‚¬ìš©. 
        self.guessed_letters = set() 
        self.lives_left = 6 
        
        
    def make_guess(self, guess):
        
        # ì›ë˜ëŠ” ì´ë¯¸ ì¶”ì¸¡í–ˆëŠ”ì§€ë¥¼ ë‘ ë²ˆì§¸ë¡œ ë’€ì—ˆëŠ”ë° ê·¸ë ‡ê²Œ í•˜ë©´ ì´ë¯¸ ì¶”ì¸¡í•œ ê¸€ìë¥¼ "ë§ì·„ë‹¤"ë¼ê³  ë¨¼ì € í•  ìˆ˜ ìˆìŒ
        # ì¶”ì¸¡í–ˆëŠ”ì§€ ë¨¼ì € ê²€ì‚¬í•˜ê³ , ê·¸ê²Œ ì•„ë‹ˆë¼ë©´ ë‹¨ì–´ë‘ ë¹„êµí•˜ê³ , ë§ê³ /í‹€ë¦¼ì„ í™•ì¸í•´ì•¼ í•¨. 
        if guess in self.guessed_letters: 
            print("ì´ë¯¸ ì¶”ì¸¡í•œ ê¸€ìì…ë‹ˆë‹¤.")

        elif guess in self.word: 
            # appendëŠ” ë¦¬ìŠ¤íŠ¸ì—ì„œ addëŠ” í•¨ìˆ˜ì—ì„œ ì”€ 
            # appendëŠ” ì¤‘ë³µì„ í—ˆìš©í•˜ê³ , addëŠ” ì¤‘ë³µì„ ë¶ˆí—ˆí•¨
            # appendëŠ” ìˆœì„œë¥¼ ìœ ì§€í•˜ê³ , addëŠ” ìˆœì„œê°€ ì—†ìŒ 
            # í–‰ë§¨ ê²Œì„ì—ì„œ addë¥¼ ì´ìš©í•˜ëŠ” ì´ìœ ëŠ” ì§‘í•©ìœ¼ë¡œ ì¤‘ë³µ ì¶”ì¶•ì„ ë°©ì§€í•˜ê¸° ìœ„í•¨
            self.guessed_letters.add(guess)
            print("ê¸€ìë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”") 
                                
        else:
            #í‹€ë ¸ì–´ë„ ì¶”ì¸¡í•œ ë ˆí„°ë¡œ ì¶”ê°€í•´ì•¼ ë‚˜ì¤‘ì— ë˜ í‹€ë¦¬ë©´ ë˜ í‹€ë ¸ë‹¤ê³  ì•Œë ¤ì¤Œ. 
            self.guessed_letters.add(guess) 
            self.lives_left -= 1
            print(f"(í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ íšŸìˆ˜ {self.lives_left})")

    def display(self):
        #resutlëŠ” ë¹ˆ ë¬¸ìì—´ë¡œ ì‹œì‘í•¨
        result = ""
        # for char in self.wordëŠ” self.wordì— ìˆëŠ” ë¬¸ìë¥¼ í•˜ë‚˜ì”© ê°€ì ¸ì˜´. 
        # ê° characterë¥¼ ìˆœíšŒí•˜ë¼ëŠ” ëœ» 
        # ê° characterë¥¼ ìˆœíšŒí•´ì„œ self.guessed_lettersê°€ self.wordì— ìˆìœ¼ë©´ 
        # result charë¥¼ ì¶”ê°€í•˜ê³  í• ë‹¹í•˜ë¼ëŠ” ëœ». 
        # ë§Œì•½ì— ì—†ìœ¼ë©´ resultì— "-"ë¥¼ ì¶”ê°€í•˜ê³  í• ë‹¹í•˜ë¼ëŠ” ëœ» 
        for char in self.word:
            if char in self.guessed_letters:
                result += char
            else: 
                result += "_"
        return result
    
    # """"""ë¡œ ë¬¶ì—¬ ìˆëŠ” í•˜ë‚˜ê°€ ì¸ë±ìŠ¤ í•˜ë‚˜ë‹ˆê¹Œ... ê°ê° ì¸ë±ìŠ¤ë¥¼ ëª©ìˆ¨ì´ ì‚¬ë¼ì§€ë©´ì„œ ë³´ì—¬ì£¼ë¼ëŠ” ëœ». 
    def display_hangman(self):
        index = 6 - self.lives_left
        return hangman_pics[index]     
    
    # play ë©”ì„œë“œ. 
    # ë¨¼ì € í–‰ë§¨ ê²Œì„ì„ ì‹œì‘í•œë‹¤ëŠ” ë©”ì‹œì§€ë¥¼ ë³´ì—¬ì£¼ê³  
    # ë‹¨ì–´ ëª‡ê°œ ë§ì·„ëŠ”ì§€ self.display ë©”ì„œë“œ ë³´ì—¬ì¤Œ 
    # ê·¸ë¦¬ê³  self.display_hangman ë„ ë³´ì—¬ì¤Œ. 

    def play(self):
        print("í–‰ë§¨ ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤")
        print(f"ë‹¨ì–´: {self.display()}")
        print(self.display_hangman())

     # while self.lives_leftê°€ 0ë³´ë‹¤ í°ë° self.guessed_lettersì— ìˆëŠ” characterê°€ self.word characterì— ëª¨ë‘ ìˆìœ¼ë©´ 
    # ì¶•í•˜í•œë‹¤ê³  ë§í•´ì£¼ê³  ê²Œì„ ì¢…ë£Œ 


        while self.lives_left > 0:
            if all(char in self.guessed_letters for char in self.word):
                print(f"ì¶•í•˜í•©ë‹ˆë‹¤. ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤:{self.word}")
                return
    
    # ê·¸ê²Œ ì•„ë‹ˆë©´ ê³„ì† ê¸€ìë¥¼ ì…ë ¥í•˜ë¼ê³  ë³´ì—¬ì£¼ë©´ ë¨ (ëŒ€ì†Œë¬¸ì ì˜¤ë¥˜ë°©ì§€ ì¶”ê°€)
    # ê·¸ë¦¬ê³  make_guess ë©”ì„œë“œ ê³„ì† ëŒì•„ê°... inputê³¼ í•¨ê»˜ 
            guess = input("ê¸€ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: ").lower().strip()
            if not guess: 
                print("ê¸€ìë¥¼ ì…ë ¥í•˜ì„¸ìš”!")
                continue

            self.make_guess(guess)

    # ê·¸ë¦¬ê³  ê³„ì† ë‹¨ì–´ ìƒí™©íŒê³¼ í–‰ë§¨ ê·¸ë¦¼ ê³„ì† ë³´ì—¬ì£¼ê³  
    # ì¶”ì¸¡í•œ ê¸€ìëŠ” ì½¤ë§ˆë¡œ ë¶„ë¥˜í•˜ê³  ì•ŒíŒŒë²³ ìˆœìœ¼ë¡œ ì •ë ¬í•´ì„œ ì¶”ì¸¡í•œ ê¸€ìë¼ê³  ê³„ì† ë³´ì—¬ ì¤Œ. 
            print(f"ë‹¨ì–´:{self.display()}")
            print(self.display_hangman())
            print(f"ì¶”ì¸¡í•œ ê¸€ì:{','.join(sorted(self.guessed_letters))}")
            print("_" * 30)

        print(f"ê²Œì„ì˜¤ë²„. ì •ë‹µì€ {self.word}ì˜€ìŠµë‹ˆë‹¤.")


if __name__ == "__main__":
    game = HangmanGame(words)
    game.play()
```
